```
import pandas as pd

# Set Pandas display options
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 20)
pd.set_option('display.width', None)
pd.set_option('display.expand_frame_repr', False)   # Disable wrapping

# Read in Flowster and Amazon datasets as Pandas dataframes
flowsterDF = pd.read_csv('/content/drive/My Drive/Github/mlteam4/team_flowster_work/teamflowster_webscraping/datasets/Flowster_Topic_Attributes_20200609181520.csv')
amazonDF = pd.read_csv('/content/drive/My Drive/Github/mlteam4/team_flowster_work/teamflowster_webscraping/datasets/amazon_scraped_data.csv')

# Extract desired Flowster columuns and remove the last row since the category only has one topic
newFlowsterDF = flowsterDF[['Topic Title', 'Category', 'Author', 'Leading Comment', 'Other Comments']]
newFlowsterDF = newFlowsterDF.drop(index=260)
#print(newFlowsterDF)

# Extract desired Amazon columuns and rename columns
newAmazonDF = amazonDF[['Title', 'Category', 'Post Author', 'Leading Comment', 'Reply Comments']]
newAmazonDF = newAmazonDF.rename(columns={'Title':'Topic Title', 'Post Author':'Author', 'Reply Comments':'Other Comments'})
#print(newAmazonDF)

# Merge new dataframes and save as CSV file
mergedDF = newFlowsterDF.append(newAmazonDF, ignore_index=True)
#print(mergedDF)
mergedDF.to_csv('/content/drive/My Drive/Github/mlteam4/team_flowster_work/teamflowster_webscraping/datasets/Amazon_Flowster_Merged_Data.csv')
```
