```python
import pandas as pd

#read data from both forums
amazon_data = pd.read_csv('amazon_scraped_data.csv')
amazon_data = amazon_data[['Title', 'Post Author', 'Leading Comment', 'Reply Comments', 'Category', 'Forum']]
flowster_data = pd.read_csv('flowster_scraped_data.csv')
flowster_data = flowster_data[['Title', 'Post Author', 'Leading Comment', 'Reply Comments', 'Category', 'Forum']]
```

```python
#joining the two dataframes
merged = pd.concat([amazon_data, flowster_data], join='inner')
merged
```

```python
#minimal data cleaning
import re
from html.parser import HTMLParser

def text_cleaning(text):
    # converting HTML character codes to ASCII code
    parser = HTMLParser()
    text = parser.unescape(text)

    text = re.sub(r'<[^>]+>', '', text)  # removing HTML tags
    text = re.sub(r'(?:\#+[\w_]+[\w\'_\-]*[\w_]+)', '', text)  # removing hash-tags
    text = re.sub('\n', ' ', text)  # remove new line
    text = re.sub('@', '', text)  # remove @ sign
    #text = re.sub(r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', '',text)  # removing URLs
    text = re.sub(r'(?:[\ufffd]+)', '', text)  # removing special characters

    return text

merged['Post Author'] = merged['Post Author'].apply(lambda x: text_cleaning(x))
merged['Title'] = merged['Title'].apply(lambda x: text_cleaning(x))
merged['Leading Comment'] = merged['Leading Comment'].apply(lambda x: text_cleaning(str(x)))
merged['Reply Comments'] = merged['Reply Comments'].apply(lambda x: text_cleaning(x))
merged['Forum'] = merged['Forum'].apply(lambda x: text_cleaning(x))
merged['Category'] = merged['Category'].apply(lambda x: text_cleaning(x))
merged
```

```python
merged.to_csv('final_merged_data.csv')
```
