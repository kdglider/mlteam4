```python
import requests
from bs4 import BeautifulSoup
from requests import get
```

```python
# this is an example of scraping a comment on the flowster forum
url= 'https://forum.flowster.app/t/about-the-amazon-specific-category/35'
```

```python
#import the parsers
response= requests.get(url)
soup= BeautifulSoup(response.text,'html.parser')
```

```python
soup.find_all('span')
```

```python
soup.find_all('div')
```

```python
soup.find_all('h3')
```

```python
soup.find_all('a')
```

```python
#once you run these commands u get the tags associated and the names for all tags and when you run thru the parser make sure
#to understand which class or parser to put thru to find the thing u want 


#for example to print category name in all the a tags extracted finding the href link returns u the name which is associated with it
```

```python
#this is to print the category name of the comment
categoryname= soup.find_all('a',href="/t/about-the-amazon-specific-category/35")
categoryname
```

```python
#from the parsed category now u extract the category name
for n in categoryname:
    for a in n:
        print(a)
```

```python
#this is to parse the category name 
comment= soup.find_all('div',class_="post")
comment
```

```python
#now extracting the comment 
for comm in comment:
    for c in comm:
        for a in c:
            print(a)
```
