```python
import requests
from bs4 import BeautifulSoup
import ssl
import urllib
```

```python
# Find a website to scrape and we save the link in that form
url= 'https://www.discoursehub.com/communities/'
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
#url = "https://www.discoursehub.com/communities/"
html = urllib.request.urlopen(url, context = ctx).read()
print(html)
soup = BeautifulSoup(html, 'html.parser')
```

```python
#Requesting a website 
page= requests.get(url)
```

```python
#parsing all categories from the given url
categorytags= soup.find_all('h2')
for categories in categorytags:
    print(categories.get_text())
```

```python
#extracting all links from a website
soup.a.get('href')
```

```python
ta= soup.find_all('a',class_= "button")
urllist= []
for tags in ta:
    #print(tags.get('href'))
    urllist.append(tags.get('href'))
print(urllist)
```

#### Let us scrape the AI forum

```python
#https://community.mycroft.ai/c/general/5
```

```python
newurl='https://community.mycroft.ai/c/general/5'
```

```python
samplepage= requests.get(newurl)
```

```python
sampleparser= BeautifulSoup(samplepage.text, 'html.parser')
```

```python
commenturllist=[]
commentlinks= sampleparser.find_all('a',class_="title raw-link raw-topic-link")
for commentlink in commentlinks:
    #print(commentlink.get('href'))
    commenturllist.append(commentlink.get('href'))
print(commenturllist)
```
